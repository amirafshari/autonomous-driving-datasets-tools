```python
import os
import re
import numpy as np
import cv2
```

```python
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts
```

```python
names = []
files = []
for e in sorted(os.listdir('./data/label'), key=numericalSort):
    names.append(e)
    with open(os.path.join('./data/label', e)) as f:
        files.append(f.read().split())


# create a dict with file names and values        
dic = {}
for n, f in zip(names, files):
    dic[n] = f


# remove classes    
for key, value in dic.items():
    l = []
    for v in value:
        if len(v) != 1:
            l.append(v.split(','))
    dic[key] = l

    
    
# remove repeated elements   
indexes = [3, 4, 6, 7]
for value in dic.values():
    for element in value:
        for index in sorted(indexes, reverse=True):
            del element[index]
            

            
            
# turn strings to integers
for objects in dic.values():
    for obj in objects:
        for i, e in enumerate(obj):
            obj[i] = int(e)


            
# read images and save their height and width in a new list
im_names = []
sizes = []
for e in sorted(os.listdir('./data/img'), key=numericalSort):
    im_names.append(e)
    im = cv2.imread(f'./data/img/{e}')
    h, w, c = im.shape
    sizes.append([h,w])
    
```

```python
for objects, size in zip(dic.values(), sizes):
    for obj in objects:
        x_center = (obj[0] + obj[2]) / (2*size[1]) # xmax + xmin
        y_center = (obj[1] + obj[3]) / (2*size[0]) # ymax + ymin
        obj_width = (obj[2] - obj[0]) / size[1]
        obj_height = (obj[3] - obj[1]) / size[0]
        
        obj.insert(0,0)
        obj[1] = x_center
        obj[2] = y_center
        obj[3] = obj_width
        obj[4] = obj_height
```

```python
fmt = '%.10g'

for n, f in zip(names, dic.values()):
    print(f)
    for e in f:
        np.savetxt(f'./data/txt/{n}', f, newline='\n',fmt=fmt)
```
